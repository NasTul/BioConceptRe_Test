```python
#read ann file
def ann_strtonum_feed(filename):
    data_list = []
    with open(filename, 'r', encoding='UTF-8') as f:#with auto call close()
        line = f.readline()
    
        while line:
            read_data = line.split('\n')[0].split('\t')
            line = f.readline()
            if read_data=="":
                continue
            number = read_data[0]
            mid_data =  read_data[1].split(" ")
                
            #handle when there has more than one label in one line
            if ';' in read_data[1]:
                continue
                
            else:
                
                name = mid_data[0]
                start = mid_data[1]
                end = mid_data[2]
                sub_content = read_data[2].split()
                for content in sub_content:
                    if data_list:
                        if  start==data_list[-1][2] and  content==data_list[-1][4]:
                            continue
                        if start != data_list[-1][2] and int(start) <  int(data_list[-1][3]) :
                            continue
                        if start == data_list[-1][2] and int(end) !=  int(data_list[-1][3]) :
                            continue
                    data_list.append((number,name,start,end,content))
            
            
            
        return data_list
```

```python




def func_handle_txt(ann_data, sentence_list):
    index = 0
    handle_sentence_list=[]
    record = 0
    for sub_data in sentence_list:
        sub_sent = []
        if index >= len(ann_data):
            break
        for i_number in range(len(sub_data)) :
            record+=2
            label ='O'
            if ann_data[index][4] in  sub_data[i_number] or sub_data[i_number]+"." in ann_data[index][4] :
                if ann_data[index-1][1] == ann_data[index][1] and  ann_data[index-1][2] == ann_data[index][2] and  ann_data[index-1][3] == ann_data[index][3]: 
                    label='I'
                else:
                    label='B'
                index +=1

            sub_sent.append((sub_data[i_number],label))
            if index >= len(ann_data):
                break
        handle_sentence_list.append(sub_sent)
    return handle_sentence_list
```

```python
all_data =[]
for i in doct_ids:
    sentence_list = func_read_txt(i[1])
    ann_data = ann_strtonum_feed(i[0])
    all_data = all_data + func_handle_txt(ann_data,sentence_list)
```
